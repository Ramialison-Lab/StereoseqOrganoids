```python
# Create figure and axes with broken Y-axis
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(12, 6),
                               gridspec_kw={'height_ratios': [1, 1]})
fig.subplots_adjust(hspace=0.05)  # Adjust space between axes

# Plot Seaborn barplot directly on Matplotlib axes
sns.barplot(x='sample', y='percentage', data=dfhouse,hue='Housekeeping gene',palette=['#E8C5E5','#91DDCF'],
            order=dfGAPDH['sample'], ax=ax1)

sns.barplot(x='sample', y='percentage', data=dfhouse,hue='Housekeeping gene',palette=['#E8C5E5','#91DDCF'],
             order=dfGAPDH['sample'], ax=ax2)



# Set Y-axis limits for the broken effect
ax1.set_ylim(75, 100)  # Upper section (zoomed into large values)
ax2.set_ylim(0, 25)     # Lower section (zoomed into smaller values)

# Hide spines between the two axes
ax1.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)

# Remove x-axis ticks from the top plot (ax1)
ax1.tick_params(bottom=False, labelbottom=False)  # No ticks at the top


# Add slanted break lines
d = 0.5  # Proportion of vertical to horizontal extent of the slanted line
kwargs = dict(marker=[(-1, -d), (1, d)], markersize=12,
              linestyle="none", color='k', mec='k', mew=1, clip_on=False)

ax1.plot([0, 1], [0, 0], transform=ax1.transAxes, **kwargs)  # Upper break line
ax2.plot([0, 1], [1, 1], transform=ax2.transAxes, **kwargs)  # Lower break line



ax2.get_legend().remove() 
# Labels and formatting
ax1.set_ylabel(' ', fontsize=14, labelpad=10)
ax2.set_ylabel(' ', fontsize=14, labelpad=10)
ax1.set_xlabel(" ", fontsize=14, labelpad=10)
ax2.set_xlabel(" ", fontsize=14, labelpad=10)

# Set a single Y-axis label for both subplots
fig.text(0.075, 0.5, 'Percentage of cells (CellBin) expressing the gene', 
         va='center', rotation=90, fontsize=14)

# Rotate x-axis labels for better readability
ax2.set_xticklabels(ax2.get_xticklabels(), fontsize=12, rotation=45, ha='right')
# Add border to bars in both subplots
for bars in ax1.patches:
    bars.set_edgecolor('black')  # Border color
    bars.set_linewidth(0.5)      # Border thickness

for bars in ax2.patches:
    bars.set_edgecolor('black')
    bars.set_linewidth(0.5)


fig.canvas.draw()

for label in ax2.get_xticklabels():
    sample = label.get_text()
    if sample in sample_palettes:
        label.set_color(sample_palettes[sample])

#for label in ax2.get_xticklabels():
#    label.set_fontweight('bold')  # Make it bold

# Adjust figure size
fig.set_size_inches(15, 6)
# Add a title to the figure
fig.suptitle('Percentage of Cells Expressing GAPDH and ACTB Per Sample', fontsize=16)

plt.savefig('/group/tran3/mnucera/STOmics_crossorganoid_project/STOmics_notebooks/Figure1_stomics/FIG1_G_GAPDH_ACTB.pdf',dpi=300)
plt.savefig('/group/tran3/mnucera/STOmics_crossorganoid_project/STOmics_notebooks/Figure1_stomics/FIG1_G_GAPDH_ACTB.svg',dpi=300)
# Display the plot
plt.show()
```
